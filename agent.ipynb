```python
import os, getpass
from openai import OpenAI


base_url = 'https://llm.keyvalue.systems'
api_key = '---'

MODEL = "openai/gpt-4o-mini"   # the model alias your LiteLLM proxy serves

client = OpenAI(
    base_url=base_url,
    api_key=api_key,
)
print("LiteLLM client ready | base_url:", base_url, "| model:", MODEL)
```

```python
def ask(prompt, system="You are a concise and friendly assistant.",
        temperature=0.7, model=MODEL):
    r = client.chat.completions.create(
        model=model,
        temperature=temperature,
        messages=[
            {"role": "system", "content": system},
            {"role": "user", "content": prompt},
        ],
    )
    return r.choices[0].message.content, (r.usage.prompt_tokens, r.usage.completion_tokens)

print("ask() defined")
```

```python
# ============================================================
# Checkpoint A - First breath: run it, get a reply printed
# ============================================================
reply, usage = ask("What's the capital of France?")
print(reply)
print("tokens (in, out):", usage)      
```

```python
history=[]
temperature = float(input("Set temperature: "))
while True:
    prompt = input("Ask away: ")
    history.append({"role": "user", "content": prompt})
    r = client.chat.completions.create(
        model=MODEL,
        temperature=temperature,
        messages=history
    )
    response=r.choices[0].message.content
    history.append({"role": "assistant", "content": response})
    print("user: ",prompt)
    print("assostant: ",response)
```
